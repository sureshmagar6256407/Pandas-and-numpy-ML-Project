```python
# read io string file
df =pd.read_csv(io.StringIO(data)) 
```

```python
# #age fill
df.fillna(
    {"Age":df["Age"].mean()} , 
    inplace= True
)
```

```python
#experience fill with median  
df.fillna (
    {"Experience_Years":df["Experience_Years"].median()}, inplace= True
)
```

```python
# Performance_Score missing → department-wise average le fill
departmentWiseAve = df.groupby("Department")["Performance_Score"].mean()  
print(departmentWiseAve)
df["Performance_Score"] = df["Performance_Score"].fillna(df["Department"].map(departmentWiseAve))
```

```python
#Salary negative xa → absolute value banaune
df["Salary"] = df["Salary"].mask(df["Salary"]<0, np.nan)
# Salary missing xa → department-wise average le fill
salarGroupBy = df.groupby("Department")["Salary"].mean() 
df["Salary"] = df["Salary"].fillna(df["Department"].map(salarGroupBy))
```

```python
# City missing → "Unknown" le fill
df["City"] = df["City"].fillna("Unknown")
```

```python
#TotalValue column add garne (Salary * Performance_Score)
df["TotalValue"] = df["Salary"]*df["Performance_Score"]
```

```python
#Experience level column add garne (Junior, Mid, Senior)
def getExp(x) : 
    if x <= 2 and x >=0 : 
        return "Junior"
    elif x<= 5 and x >2 : 
        return "Mid"
    else : 
        return "Senior"
df["ExperienceLevel"] = df["Experience_Years"].apply(getExp)
```

```python
# Joining_Date lai datetime ma convert garne
df["Joining_Date"] = pd.to_datetime(df["Joining_Date"], errors="coerce")

# Years_in_Company calculate gara 
df["yearsinCompany"] = pd.to_datetime("today").year- df["Joining_Date"].dt.year

```

```python
# Highest paid employee ko name nikalne  
highestPaidEmployee  = df.loc[df["Salary"].idxmax()]["Name"]
print(highestPaidEmployee)
```

```python
# Best performance score ko employee ko name nikal
BestperformanceEmployee = df.loc[df["Performance_Score"].idxmax()]["Name"]
print(BestperformanceEmployee)
```

```python
#Department-wise average salary  
departmentWiseAverageSalary  = df.groupby("Department")["Salary"].mean()
print(departmentWiseAverageSalary)
```

```python
# Experience level wise average performance nikal
experiencelevelwiseAvePer  = df.groupby("ExperienceLevel")["Performance_Score"].mean()
print(experiencelevelwiseAvePer)
```

```python
# Performance_Score < 75 bhako employee remove garne
df= df[df["Performance_Score"]>= 75].copy()
df.reset_index(drop=True,inplace=True)
```
