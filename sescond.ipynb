```python
#module download   
import numpy as np 
import pandas as pd 
```

```python
#data 
data = {
    "Size_sqft": [500, 800, 1000, 1200, 1500, 1800, 2000, 2200, 2500, 3000],
    "Bedrooms": [1, 2, 2, 3, 3, 3, 4, 4, 4, 5],
    "Age": [10, 8, 6, 5, 4, 3, 2, 2, 1, 1],
    "Price": [5000000, 8000000, 10000000, 12000000, 15000000, 18000000, 20000000, 22000000, 25000000, 30000000]
}
df = pd.DataFrame(data)
```

```python
#total average house price 
totalAveragePrice = np.mean(df["Price"])  
print(f"The total average house price is : {totalAveragePrice}")
```

```python
#highest house price index 
highestPriceIndex =np.argmax(df["Price"])
print(f"the highest house price index id : {highestPriceIndex}")
```
